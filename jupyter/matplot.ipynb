```python
import numpy as np  
import matplotlib.mlab as mlab  
import matplotlib.pyplot as plt  

# example data  
mu = 100 # mean of distribution  
sigma = 15 # standard deviation of distribution  
x = mu + sigma * np.random.randn(10000)  

num_bins = 50  
# the histogram of the data  
n, bins, patches = plt.hist(x, num_bins, normed=1, facecolor='blue', alpha=0.5)  
# add a 'best fit' line  
y = mlab.normpdf(bins, mu, sigma)  
plt.plot(bins, y, 'r--')  
plt.xlabel('Smarts')  
plt.ylabel('Probability')  
plt.title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')  

# Tweak spacing to prevent clipping of ylabel  
plt.subplots_adjust(left=0.15)  
plt.show()
```

```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
#
labels = np.array([u'3',u'5',u'6',u'3',u'1',u'3',u'3',u'2'])
#
dataLenth = 8
#
data = np.array([3,5,6,3,1,3,3,2])


angles = np.linspace(0, 2*np.pi, dataLenth, endpoint=False)
data = np.concatenate((data, [data[0]]))
angles = np.concatenate((angles, [angles[0]]))
fig = plt.figure(figsize=(11,9))
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, data, 'ro-', linewidth=2)
ax.set_thetagrids(angles * 180/np.pi, labels, fontproperties="SimHei",fontsize=20)
ax.set_title(u"", va='bottom', fontproperties="SimHei",fontsize=20)
ax.grid(True)
plt.show()
```
